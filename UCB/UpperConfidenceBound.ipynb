```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Ads_CTR_Optimisation.csv')
data.head()
```

```python
N = len(data)
d = len(data.columns)
ads_selected = []
n_selections = [0] * d
sum_of_rewards = [0] * d
total_reward = 0
```

```python
for n in range(0,N):
    ad = 0
    max_bound = 0
    for i in range(0, d):
        
        if(n_selections[i]>0):
            avg_reward = sum_of_rewards[i] / n_selections[i]
            delta_i = math.sqrt(1.5 * math.log(n+1) / n_selections[i])
            upper_bound = avg_reward + delta_i
        else:
            upper_bound = 1e400
            
        if upper_bound > max_bound:
            max_bound = upper_bound
            ad = i
    
    ads_selected.append(ad)
    n_selections[ad] = n_selections[ad] + 1
    reward = data.iloc[n, ad]
    sum_of_rewards[ad] = sum_of_rewards[ad] + reward
    total_reward = total_reward + reward
    
```

```python
total_reward
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
